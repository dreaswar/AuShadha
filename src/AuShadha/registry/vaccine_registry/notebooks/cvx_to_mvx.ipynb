```python
#! /usr/bin/python

##################################################
# Project: This code is part of AuShadha Project
# Author: Dr. Easwar T.R
# Date: 09 Aug 2014
# License: GNU-GPl Version 3 
##################################################

"""
  Parser for the CVX-MVX from CDC.
  CVX File contains the data for all the CDC approved Vaccines.
  The XML is available from CDC website. http://cdc.gov
  For usage details, license please consult the website
"""


#Imports
import os
import sys
import copy
import json
import datetime
from xml.etree import cElementTree as ET


#Constants
ROOT_PATH = os.path.abspath(os.path.curdir)
APP_PATH = os.path.abspath(os.path.pardir)

DATA_DIR = ROOT_PATH + '/data/cdc/'
JSON_FIXTURE_PATH = ROOT_PATH+'/cvx_to_mvx_initial_data.json'
APP_FIXTURE_PATH = APP_PATH+'/fixtures/initial_data.json'

#Parse the XML
CVX_XML = ET.parse(DATA_DIR+"/cvx_to_mvx.xml") 
CVX_ROOT = CVX_XML.getroot()

#Basic parsed info
print("Parsed the CVX-MVX XML")
print("ROOT tag: " + CVX_ROOT.tag)

#Store vaccine info as iterable to be used later
all_vaccines = (lambda: [v for v in CVX_ROOT.findall('prodInfo')])() # Store all the Vaccines in a iterable form

def to_django_fixture(format = 'json'):
    v_count = 0
    data = []
    MODEL_DICT = {'pk': '', 
                  'fields': {},
                  'model': 'vaccine_registry.VaccineCVXToMVX'
                  }
    
    for _v in all_vaccines:
        _name = [_.text for _ in _v.findall('Name')]
        _value= [_.text for _ in _v.findall('Value')]
        _M = copy.deepcopy(MODEL_DICT)
        _M['pk']= v_count
        def update_model_fixture(k,v):
            k = k.replace(' ','_').lower()
            if v: 
                v = v.strip()
                if k == 'cvxcode': 
                    v = int(v)
                elif k == 'last_updated':
                    v = datetime.datetime.strptime(v,'%m/%d/%Y').isoformat()
            _M['fields'][k] = v
        map(update_model_fixture,_name, _value)
        data.append(_M)
        v_count +=1

    out_data = json.dumps(data,indent = 4)
    with open(JSON_FIXTURE_PATH,'w') as out_file:
        out_file.write(out_data)
        out_file.close()
        print("Fixture Data Written")
    with open(APP_FIXTURE_PATH,'a') as out_file:
        out_file.write(out_data)
        out_file.close()
        print("App Fixture Updated")    
    print(out_data)


to_django_fixture()
```
